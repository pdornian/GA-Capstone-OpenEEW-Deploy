# Map Generation For Predictions

Below, we're going to write a function that given the index of an earthquake in our test set returns a map that plots the earthquake and the predicted location.

```python
preds_df=pd.read_pickle("..\..\data\\pickled_files\predictions.p")
```

```python
preds_df
```

Also, let's reset the index and sort by magnitude.

```python
preds_df=preds_df.sort_values("magnitude", ascending=False)
```

```python
preds_df.reset_index(drop=True, inplace=True)
```

```python
preds_df
```

Let's also make a distance column that tracks how far off our prediction is.

```python
def get_distance(df_row):
    actual=(df_row['latitude'], df_row['longitude'])
    pred = (df_row['pred_lat'], df_row['pred_long'])
    return float(distance.distance(actual, pred).km)
```

```python
preds_df['error']=preds_df.apply(get_distance, axis=1)
```

```python
preds_df.sort_values('magnitude', ascending=False, inplace=True)
```

```python
pickle.dump(preds_df, open("..\..\data\\pickled_files\predictions.p", "wb"))
```

```python
def map_quake(index, df):
    quake=df.loc[index, :]
    lat=quake['latitude']
    long=quake['longitude']
    pred_lat=quake['pred_lat']
    pred_long=quake['pred_long']
    m = folium.Map(location=[(lat + pred_lat)/2, (long + pred_long)/2], zoom_start=7)
    actual = folium.Circle(radius=10_000,
                           location=[lat, long],
                           color='black',
                           fill=True,
                           fill_color='orange',
                           fill_opacity=1,
                           popup=folium.Popup("Earthquake Location", max_width=200))
    pred = folium.Circle(radius=10_000,
                         location=[pred_lat, pred_long],
                         color='black',
                         fill=True,
                         fill_color='green',
                         fill_opacity=1,
                         popup=folium.Popup(f'Predicted Location \n Error: {int(round(quake["error"],0))} km', max_width=100)
                         )
    line= folium.PolyLine([(lat, long), (pred_lat, pred_long)],
                          color="black",
                          dash_array="4",
                          weight=1)
    line.add_to(m)
    actual.add_to(m)
    pred.add_to(m)
    return m
    
```

```python
map_quake(106, preds_df)
```
