```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, stratify= fucked_df['cluster'])
```

```python
X_train.shape
```

```python
X_train= np.array(X_train)
X_test= np.array(X_test)
y_train= np.array(y_train)
y_test= np.array(y_test)
```

```python
sample_row= X_train[0, :]
```

```python
test=np.dstack(sample_row)
```

```python
def make_arrays(df_slice):
    convert = np.array(df_slice)
    output=[]
    i=0
    for row in convert:
        array = np.dstack(row)
        output.append(array)
        i+=1
    return output
```

```python
X_train_arr = np.array(make_arrays(X_train))
```

```python
X_train_arr[0].shape
```

```python
X_test_arr = np.array(make_arrays(X_test))
```

```python
model=Sequential()
model.add(Dense(32, activation='relu'))
model.add(Flatten())
model.add(Dense(16, activation='relu'))
model.add(Flatten())
model.add(Dense(2, activation='linear'))
```

```python
model(expand_dims(X_train_arr[0], axis=0))
```

```python
model.summary()
```

```python
model.compile(loss = 'mse', optimizer='adam', metrics=['mse'])
```

```python
len(X_train_arr)
```

```python
y_train.shape
```

```python
model.fit(X_train_arr, y_train, validation_data=(X_test_arr, y_test), epochs=100, batch_size=20)
```

```python
train_preds=model.predict(X_train_arr)
test_preds=model.predict(X_test_arr)
```
