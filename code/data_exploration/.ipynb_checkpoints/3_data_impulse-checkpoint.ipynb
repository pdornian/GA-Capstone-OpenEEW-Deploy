```python
import pandas as pd
import numpy as np
import pickle
from tqdm import tqdm
```

The following worksheet was created after the first round of modelling from the 5 minute waveforms didn't turn up any useful results.

-----------------------------------------------------------

I think the waveforms I put in were too long and don't say anything. Also, interpolating missing data to be 0 matrices might have been A Bad Idea.

To see if there's anything I can extract from them (and there might not be), I'm going to try two different metrics.

**Most Active 30 Seconds**: We're going to toss out the consistent timeframe, and just have each device report the most active 30 seconds of data. This will reduce the inactive noise.

**Peak Impulse**: Collapse the 30s waveform into the sum of the absolute values of it's entries to associate a scalar to each device for a given quake. Once we do that, we're  working with vector data again and can whip out the sklearn toolkit to try to identify things.

```python
final_df=pd.read_pickle("top_1000_final.p")
```

```python
device_cols=final_df.columns[final_df.columns.str.startswith('device')]
```

```python
#returns a (waveform, int) tuple of the most active 30s and the max impulse
def max_30s(waveform):
    max_impulse=0
    max_sub_wf=np.zeros((3,30))
    for i in range(271):
        sub_wf=waveform[:,i:i+30]
        impulse = np.sum(np.abs(sub_wf))
        if impulse > max_impulse:
            max_impulse=impulse
            max_sub_wf=sub_wf
    return (max_sub_wf, max_impulse)
```

```python
wf_df=final_df.copy()
impulse_df=final_df.copy()
```

```python
for device in tqdm(device_cols):
    wf_df[device]=wf_df[device].map(lambda x: max_30s(x)[0])
    impulse_df[device]=impulse_df[device].map(lambda x: max_30s(x)[1])
```

```python
pickle.dump(impulse_df, open('impulse_df.p', "wb"))
pickle.dump(wf_df, open('wf_df.p', "wb"))
```

```python
impulse_df=pd.read_pickle('impulse_df.p')
wf_df=pd.read_pickle('wf_df.p')
```
